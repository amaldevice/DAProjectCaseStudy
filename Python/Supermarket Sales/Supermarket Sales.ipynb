```python
branch = df.groupby(by='Branch').agg({
    'Quantity':['sum']
})
branch
```

# Data Visualization

```python
branch.plot.pie(subplots=True, autopct='%1.1f%%', legend=False, labels=None)

# Menambahkan label pada potongan-potongan pie chart
plt.legend(branch.index, title='Cabang', bbox_to_anchor=(1.0, 0.5))

plt.title('Distribusi Jumlah Barang per Cabang')
plt.axis('equal')  # Membuat lingkaran terlihat seperti lingkaran sejati
plt.show()
```

## Total penjualan barang percabang

```python
city = df.groupby(by='City').agg({
    'Total':['sum']
})
city
```

```python
city_sorted = city.sort_values(by=('Total', 'sum'), ascending=False)
ax = city_sorted.plot(kind='bar', legend=False)
plt.title('Total Penjualan per Kota (Diurutkan dari Teratas)')
plt.xlabel('Kota')
plt.ylabel('Total Penjualan')
for index, value in enumerate(city_sorted[('Total', 'sum')]):
    ax.text(index, value, f'{value:.1f}', ha='center', va='bottom')
plt.xticks(range(len(city_sorted.index)), city_sorted.index, rotation=45)
plt.show()
```

## Total penjualan per kota

```python
total = df.groupby(by='Customer type').agg({
    'Total':['sum']
})
total
```

```python
total.plot.pie(subplots=True, autopct='%1.1f%%', legend=False, labels=None)
plt.legend(total.index, title='Jenis Pelanggan', bbox_to_anchor=(1.0, 0.5))
plt.title('Distribusi Jenis Pelanggan')
plt.axis('equal')  
plt.show()
```

## Total penjualan berdasarkan gender

```python
category = df.groupby(by='Product line').agg({
    'Quantity':['sum']
})
category
```

```python
category_sorted = category.sort_values(by=('Quantity', 'sum'), ascending=False)
ax = category_sorted.plot(kind='bar', legend=False)
plt.title('Total Quantity per Product Line')
plt.xlabel('Product Line')
plt.ylabel('Total Quantity')
plt.xticks(range(len(category_sorted.index)), category_sorted.index, rotation=45)  # Menambahkan label pada batang grafik
plt.show()
```

## Total penjualan berdasarkan kategori produk

```python
payment = df.groupby(by='Payment').agg({
    'Invoice ID':['count']
})
payment
```

```python
payment.plot.pie(subplots=True, autopct='%1.1f%%', legend=False, labels=None)
plt.legend(payment.index, title='Metode Pembayaran', bbox_to_anchor=(1.0, 0.5))
plt.title('Distribusi Pembayaran')
plt.axis('equal')  # Membuat lingkaran terlihat seperti lingkaran sejati
plt.show()
```
