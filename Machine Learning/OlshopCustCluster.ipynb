```python
categorical_columns = df[['Gender', 'N_Purchases', 'Pay_Method', 'Browser', 'Newsletter', 'Voucher']]
numerical_columns = df[['Age', 'Revenue_Total', 'Purchase_VALUE', 'Time_Spent']]
```

```python
for column in numerical_columns:
    plt.figure(figsize=(12,8))
    sns.histplot(x=column, data=df)
    plt.title(f'Boxplot for {column}')
    plt.show()
```

```python
for column in numerical_columns:
    plt.figure(figsize=(12,8))
    sns.boxplot(x=column, data=df)
    plt.title(f'Boxplot for {column}')
    plt.show()
```

```python
for column in categorical_columns:
    plt.figure(figsize=(12,8))
    sns.countplot(x=column, data=df)
    plt.title(f'Countplot for {column}')
    plt.show()
```

```python
X = df.drop(['Customer_id', 'Purchase_DATE'], axis=1)
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11): 
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Metode Elbow')
plt.xlabel('Jumlah clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=0)
kmeans.fit(X)
```

```python
df['Cluster'] = kmeans.predict(X)
df
```

```python
sns.scatterplot(data=df, x='Time_Spent', y='Purchase_VALUE', hue='Cluster', palette='Set1')
plt.title('Hasil Clustering dengan K-Means (3 Kluster)')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
score = silhouette_score(X, kmeans.labels_)
print('Silhouette Score:', score)
```
