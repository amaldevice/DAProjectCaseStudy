# Data Import / Data Wrangling


This data is taken from kaggle in 4 csv table format and combined in Google BigQuery SQL. In Python i will do several stages of transformation and data cleaning.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/bq-results-20230901-162707-1693587876343.csv")
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
df.duplicated().sum()
```

```python
df.isna().sum()
```

```python
df['created_at'] = pd.to_datetime(df['created_at'])
df['paid_at'] = pd.to_datetime(df['paid_at'])
df['delivery_at'] = pd.to_datetime(df['delivery_at'])
df.info()
```

```python
unique_years = df['created_at'].dt.year.unique()
print(unique_years)
unique_month = df['created_at'].dt.month.unique()
print(unique_month)
```

```python
df.drop(['kodepos', 'order_id_1', 'product_id_1', 'user_id', 'kodepos_1'], axis=1, inplace=True)
df.info()
```

```python
df.drop(['base_price', 'order_detail_id'], axis=1, inplace=True)
df.info()
```

```python
df['delivery_time'] = df['delivery_at'] - df['paid_at']
df.info()
```

```python
df.dropna(subset=['paid_at'], inplace=True)
df.info()
```

```python
average_delivery_time = df['delivery_time'].mean()

print("Average Delivery Time:", average_delivery_time)
```

# Exploratory Data Analysis

```python
df.head()
```

```python
df.describe(include="all")
```

```python
topcategory = df.groupby(by='category').agg({
    'quantity':['count','sum'],
    'total':'sum'
})
topcategorysorted = topcategory.sort_values(by=('total', 'sum'), ascending=False)

# Membatasi hasil untuk menampilkan 10 produk tertinggi
top_category = topcategorysorted.head(10)

print(top_category)
```

```python
df.groupby(by='seller_id').agg({
    'delivery_time':['count','mean']
})
```

```python
df.resample(rule='M', on='paid_at').agg({
    'quantity':['count','sum'],
    'total':'sum'
})
```

```python
top10product = df.groupby(by='desc_product').agg({
    'quantity': ['count', 'sum'],
    'total': 'sum'
})

# Mengurutkan hasil agregasi berdasarkan jumlah total dalam urutan menurun
top10productsorted = top10product.sort_values(by=('total', 'sum'), ascending=False)

# Membatasi hasil untuk menampilkan 10 produk tertinggi
top_10_products = top10productsorted.head(10)

print(top_10_products)
```

```python
top10buyer = df.groupby(by='nama_user').agg({
    'quantity': ['count', 'sum'],
    'total': 'sum'
})

# Mengurutkan hasil agregasi berdasarkan jumlah total dalam urutan menurun
top10buyersorted = top10buyer.sort_values(by=('total', 'sum'), ascending=False)

# Membatasi hasil untuk menampilkan 10 produk tertinggi
top_10_buyer = top10buyersorted.head(10)

print(top_10_buyer)
```
